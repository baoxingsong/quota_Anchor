```python
#读取各个DEGs文件

import os
import pandas as pd
PATH = ["/home/song/project/2_CrossRnaSeq/2_data/maizeDEGs/", "/home/song/project/2_CrossRnaSeq/2_data/sorghumDEGs/"]
SPECIES = ['maize','sorghum']
TREATMENT = ['05h','1h','3h','6h','16h','24h']
keep_cols = ["Row.names", "DE_group"]

# MAIZE_CATEGORIES = ["maize_05hDEGS_coldvsCK.csv", "maize_1hDEGS_coldvsCK.csv", "maize_3hDEGS_coldvsCK.csv", "maize_6hDEGS_coldvsCK.csv", "maize_16hDEGS_coldvsCK.csv", "maize_24hDEGS_coldvsCK.csv"]
# SORGHUM_CATEGORIES = ["sorghum_05hDEGS_coldvsCK.csv", "sorghum_1hDEGS_coldvsCK.csv", "sorghum_3hDEGS_coldvsCK.csv", "sorghum_6hDEGS_coldvsCK.csv", "sorghum_16hDEGS_coldvsCK.csv", "sorghum_24hDEGS_coldvsCK.csv"]
```

```python
dfs = {}

for path in PATH:
    for treatment in TREATMENT:
        species = path.split('/')[-2][:-4]   #对文件路径进行处理获取物种
        file_path = os.path.join(path, species + '_' + treatment + 'DEGS_coldvsCK.csv')  #拼接得到文件路径
        df_name = str(species + '_' + treatment)  #批量命名数据框
        df = pd.read_csv(file_path, sep = ',', header=0, index_col=False)[keep_cols].rename(columns={"Row.names":"gene", "DE_group":"label"})  #选取特定的数据列并重命名
        df['species'] = species
        df['treatment'] = treatment
        dfs[df_name] = df
```

```python
#由于本次两文件格式差异较大，故不用循环来读入
maize_file_path = '/home/song/project/2_CrossRnaSeq/2_data/PRJNA344653_maize_tpm.1000promoter.csv'
maize_promoter = pd.read_csv(maize_file_path, sep=',', header=0, usecols=['gene','X1000promoter'])  #读取数据时仅选取需要的列
maize_promoter = maize_promoter.rename(columns={'X1000promoter':'seq'}) 

sorghum_file_path = '/home/song/project/2_CrossRnaSeq/2_data/sorghum.1000.promoters.fa.csv'
sorghum_promoter = pd.read_csv(sorghum_file_path, sep=',' ,names=['gene','seq'])  #读取数据时仅选取需要的列

#将多列表作为字典方便后续调用
promoter = {}  

promoter['maize'] = maize_promoter
promoter['sorghum'] = sorghum_promoter
```

```python
dataset = pd.DataFrame()  #创建空的数据框以保存数据

for key in dfs.keys():  #根据字典key来遍历字典
    species = key.split('_')[0]  #对key进行分割已得到其包含的物种及处理信息
    treatment = key.split('_')[1]

    df = pd.merge(dfs[key], promoter[species], how='outer', on='gene' )  #将序列和基因按基因名并集连接并找出非差异基因
    df.fillna({'species':species,'treatment':treatment,'label':'None-regulated'},inplace=True) #按列填充na值并直接将改变应用于原数据框
    df['label'] = df['label'].replace({'Up-regulated': 0, 'Down-regulated': 1, 'None-regulated':2})  #为label进行重命名
    
    order = ['gene','seq','species','treatment', 'label'] 
    df = df[order] #对数据框重新排序

    dataset = dataset.append(df) #将处理好的数据框依次填入到整体数据框中
```

```python
dataset['label'].value_counts()
```

```python
dataset.to_csv('maize_sorghum_DEGs_promoter.csv', sep=',') #将整合好的文件保存为csv文件
```

```python
# 填充序列到指定长度
def pad_sequences_to_length(seq, threshold):
    if len(seq) < threshold:
        padding_length = threshold - len(seq)
        padded_seq = seq + 'N' * padding_length
    else:
        padded_seq = seq

    return padded_seq
```

```python
threshold = 2500  #设定cds长度阈值以优化数据结构

#由于本次两文件格式差异较大，故不用循环来读入
maize_file_path = '/home/song/project/2_CrossRnaSeq/2_data/maize_cold_tpm_cds.csv'
maize_cds = pd.read_csv(maize_file_path, sep=' ', header=0, usecols=['gene','seq'])  #读取数据时仅选取需要的列

maize_cds = maize_cds[maize_cds['seq'].apply(len) <= threshold]  #筛选掉过长的序列
maize_cds['seq'] = maize_cds['seq'].apply(pad_sequences_to_length, threshold=2500)  #填充序列至指定长度

sorghum_file_path = '/home/song/project/2_CrossRnaSeq/2_data/sorghum.cds.csv'
sorghum_cds = pd.read_csv(sorghum_file_path, sep=',' , header = None, names=['gene','seq'])  #读取数据时仅选取需要的列

sorghum_cds = sorghum_cds[sorghum_cds['seq'].apply(len) <= threshold]  #筛选掉过长的序列
sorghum_cds['seq'] = sorghum_cds['seq'].apply(pad_sequences_to_length, threshold=2500)

#将多列表作为字典方便后续调用
cds_2500 = {}  

cds_2500['maize'] = maize_cds
cds_2500['sorghum'] = sorghum_cds
```

```python
dataset = pd.DataFrame()  #创建空的数据框以保存数据

for key in dfs.keys():  #根据字典key来遍历字典
    species = key.split('_')[0]  #对key进行分割已得到其包含的物种及处理信息
    treatment = key.split('_')[1]

    df = pd.merge(dfs[key], cds_2500[species], how='outer', on='gene' )  #将序列和基因按基因名并集连接并找出非差异基因
    df.fillna({'species':species,'treatment':treatment,'label':'None-regulated'},inplace=True) #按列填充na值并直接将改变应用于原数据框
    df['label'] = df['label'].replace({'Up-regulated': 0, 'Down-regulated': 1, 'None-regulated':2})  #为label进行重命名
    
    order = ['gene','seq','species','treatment', 'label'] 
    df = df[order] #对数据框重新排序

    dataset = dataset.append(df) #将处理好的数据框依次填入到整体数据框中
```
